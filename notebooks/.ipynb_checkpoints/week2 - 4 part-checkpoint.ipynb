```python
"""
Created on Thu Jan 19 17:35:27 2017

@author: gsgarbi
"""

import datetime

import matplotlib.pyplot as plt
import matplotlib.colors as mpl_colors
import matplotlib.colorbar as mpl_colorbar

import netCDF4 as nc
import numpy as np
from salishsea_tools import geo_tools

import shutil

#get_ipython().magic('matplotlib inline')

%matplotlib inline


 
#step = input("Time step in minutes(ex:30): ")

#step = int(step)

# In[4]:

#End of spit
late = 49.205
lone = -123.26
#Further up spit
lats = 49.214
lons = -123.22

deep = 110
shallow = 35




# In[9]:

bathy = nc.Dataset("/home/gsgarbi/analysis-giorgio/sequential/bathy_meter_SalishSea2.nc")
model_lats = bathy.variables['nav_lat'][:]
model_lons = bathy.variables['nav_lon'][:]
depth = bathy.variables['Bathymetry'][:]


# In[ ]:

dx = 1.2
lat = late + dx * (late - lats)
lon = lone + dx * (lone - lons)
y, x = geo_tools.find_closest_model_point(lon, lat, model_lons, model_lats)
#print(depth[y, x])
#print(y, x)
#print('And in Fortran for ariane', y + 1, x + 1)
# point I used
outy = y - 1
outx = x - 1
#print('Point I used', outx, outy)

standard_months = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']


trajectories =[]


months = [["jan", 'feb', 'mar', 'apr'], 
['may', 'jun', 'jul', 'aug'], ['sep', 'oct', 'nov', 'dec']]

trajectories1 =[]

test = 2*["jan", "feb", "mar", "apr", "may", "jun"]


    

for mon_s in standard_months:
    traj = np.loadtxt("/home/gsgarbi/analysis-giorgio/new_sequential/"+mon_s+'/traj.txt')
    trajectories1.append(traj)




line0 = trajectories1[:4]

line1 = trajectories1[4:8]

line2 = trajectories1[8:12]

trajectories = []

trajectories.append(line0)
trajectories.append(line1)
trajectories.append(line2)





fig, axs = plt.subplots(3,4,figsize=(40,40))





LINES = 3
COLUMNS = 4

for j in range(LINES):
    for k in range (COLUMNS):
#Ive chosen my month
        particles = [1, 5, 9, 12]
        edit = (months[j][k], months[j][k], months[j][k])
        init = []
        for i in range(len(trajectories[j][k])):
            if trajectories[j][k][i][0] in particles:
                init.append(trajectories[j][k][i][3])
                particles.remove(trajectories[j][k][i][0])
            if particles == []:
                particles = [1, 5, 9, 12]
                break


        edit = edit + tuple(init)

        axs[j,k].set_title("""%s 1 - %s 14
4 particles starting at 7AM - 11AM - 3PM - 6PM (UTC) on %s 1
Initial depths are: %s, %s, %s, %s, respectively """ % edit)

        
        mesh = axs[j,k].pcolormesh(model_lons, model_lats, depth, cmap='winter_r')
        cb = fig.colorbar(mesh, ax=axs[j,k])
        cmap = plt.cm.plasma_r
        
        norm = mpl_colors.Normalize(vmin=deep, vmax=shallow)


        axs[j,k].set_xlim((-123.5, -123.1))
        axs[j,k].set_xticks((-123.5, -123.3, -123.1))
        axs[j,k].set_xticklabels(('123.5 W', '123.3 W', '123.1 W'))
        axs[j,k].set_xlabel('Longitude')
        axs[j,k].set_ylim((49.05, 49.35))
        axs[j,k].set_yticks((49.1, 49.2, 49.3))
        axs[j,k].set_yticklabels(('49.1 N', '49.2 N', '49.3 N'))
        axs[j,k].set_ylabel('Latitude')
        
        axs[j,k].plot(model_lons[outy, outx], model_lats[outy, outx], 'wo')
        

        for i in range(len(trajectories[j][k])):
            if trajectories[j][k][i][0] in particles:
                scaled_z = 1-(deep + trajectories[j][k][i][3]) / (deep - shallow)
                cmap = plt.cm.plasma_r
                color = cmap(scaled_z)
                axs[j,k].scatter(trajectories[j][k][i][1] , trajectories[j][k][i][2] , c=color, edgecolor = "none")

plt.tight_layout() 
```
